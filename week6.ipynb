### Practice. 由鍵盤動態加入字典元素
- 使用split()方法

```
d = dict()
valuesum = 0
while True:
    key_value = input('請輸入姓名, 年收入(萬美金)配對, 以空格區隔:')
    if key_value=='': break
    k_v_lst=key_value.split()
    value = float(k_v_lst[1])
    d[k_v_lst[0]]=value
    valuesum += value
    
print(d)
print(f'平均年收入:{valuesum/len(d):.2f}萬美金')
```

### Practice 2. 計算names名單中，每個姓氏出現的次數。

```
d = dict()
for x in names_lst:
    if x[0] in d:
        #cnt = d[x[0]]
        d[x[0]] += 1
    else:
        d[x[0]] = 1
        
print(d)
```

### 依鍵的排序印出值
* 升冪: for key in sorted(字典.keys()):
* 降冪: for key in sorted(字典.keys(), reverse=True):

```
# 升冪
for key in sorted(d4.keys()):
  print(f'{key}=>{d4[key]}')
```

```
# 降冪
for key in sorted(d4.keys(), reverse=True):
  print(f'{key}=>{d4[key]}')
```
