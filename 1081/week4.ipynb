### range
* range([起始值, ]終止值[, 遞增量]) 起始值可以不寫，預設值是0 遞增量可以不寫，預設值是1

* range(end)  預設起始值為0可以不寫, 間隔值預設為1可以不寫.
* range(start, end) 間隔值預設為1可以不寫.
* range(start, end, step)

```
print(*range(10)) # 記得加星號
print(*range(1,11))
print(*range(1, 11, 2))
print(*range(11, 1, -1))
print(*range(11, 0, -2))
```

### 利用兩層 for 迴圈列印九九乘法表。(<ninenine.py>)

```
c = int(input('column number:'))
for k in range(2, 10, c):
  for j in range(1, 10):
      for i in range(k, k+c):
          print(f'{i:>2}*{j:>2}={i*j:>2}', end='\t')
      print()
  print()
  print()
```

### break 命令: 跳出迴圈(配合if)

### continue 命令: 繼續迴圈的下個loop(配合if)

### practice 1. 
所有可被 a、b 同時整除的數，稱為 a、b 的公倍數，眾多公倍數之中最小的即為
最小公倍數。小凱靈機一動，馬上利用 for 迴圈設計這個程式，當輸入a、b 兩個
正整數後，立即可以找出其最小公倍數。(<multiple.py>)

```
a, b=(int(x) for x in eval(input('請輸入兩整數，以逗號分隔:')))
if a==0 or b==0:
  s = 0
else:
  na, nb=a, b
  if na<0: na=-a
  if nb<0: nb=-b
  na, nb = max(na,nb), min(na,nb)
  s = na
  for i in range(1, nb):
    if s//nb*nb==s:
      break
    else:
      s += na
print(f'{a}, {b}的最小公倍數為{s}')
```

### 數學上定義「n!=1*2*3*⋯*n」，所以 1!=1、2!=1*2=2、3!=1*2*3=6
* 請用 while 設計這個程式。當使用者輸入一個正整數 n 後，程式就會顯示由 1*2*3*...*n 的乘積，100! 電腦算得出來嗎？如果可以，後面有幾個 0 呢？。(<while.py>)

```
n = int(input('請輸入一正整數n:'))
res = 1
for i in range(1, n+1):
    res *= i
print(f'{n}!={res}')
```

### else 迴圈結構 (Advanced)
* 迴圈結束時執行一次的區塊

```
output=''
s=0
for i in [3, 4, 5, 8, 11, 17, 6]:
    if i%2==0:
        continue
    else:
        output += str(i) + '+'
        s += i
else:
    output=output[:-1]+ '='
    print(f'奇數和為{output}{s}')
```
