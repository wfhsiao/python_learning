### 參數預設值: 當未給該參數時會用預設值來取代。

```python
def myfunc3(n, end):
  for i in range(n):
    print('hello', end=end)
    
myfunc3(3, ' ')
```

```python
myfunc3(3) # 未給第2參數，故出錯。
```

```python
# 如何在未給第2參數時自動用預設值取代? 答: 參數=預設值
def myfunc4(e, end='\n'):
  for i in range(e):
    print('hello', end=end)
    
myfunc4(3, ' ') #有給第2參數
```

```python
myfunc4(3)  # 未給第2參數會以預設值取代
```

```python
def mysum(lst):
    s = 0
    for i in lst:
        s += i
    return s
```

```python
lst=list(range(1, 11))
lst2=list(range(1, 101))
# 使用自定函式sum
%timeit (mysum(lst), mysum(lst2))
```

```python
# 使用內建函式sum
%timeit (sum(lst), sum(lst2))
```
