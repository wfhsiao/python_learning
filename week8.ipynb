### ord 與 chr 互為反函式
- ord 傳回字元的 unicode code (萬國碼), e.g. ord('A') 是65
- chr 傳回萬國碼對應的字元, e.g. chr(65) 是 'A'

```python
c = 0
for i in range(ord('一'), ord('二')+1):
    print(f'{chr(i)}', end=' ')
    c+=1
    if c % 20==0: print()
print(f'\n\n字元"一"到"二"(含)間共有{c}個字元')
```

### 7.2.3 最大值、最小值、總和及排序
- max
- min
- sum
- sorted

```python
def mysum(lst):
    s = 0
    for i in lst:
        s += i
    return s
```

```python
lst=list(range(1, 11))
lst2=list(range(1, 101))
# 使用自定函式sum
%timeit (mysum(lst), mysum(lst2))
```

```python
# 使用內建函式sum
%timeit (sum(lst), sum(lst2))
```

```python
vs = [56, 17, 89, 94, 19, 29]
print(f'原始資料: {vs}')
vs = sorted(vs)
print(f'升冪排序: {vs}')
vs = sorted(vs, reverse=True)
print(f'降冪排序: {vs}')
```

```python
mystr='''Help on method_descriptor:

find(...)
    S.find(sub[, start[, end]]) -> int
    
    Return the lowest index in S where substring sub is found,
    such that sub is contained within S[start:end].  Optional
    arguments start and end are interpreted as in slice notation.
    
    Return -1 on failure.'''
```

```python
mstr=mystr  #'12This is a test. Is is a verb.'
substr='in'
p = mstr.find(substr)
while p != -1:
   print(p)
   p = mstr.find(substr, p+1)
```

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.managertoday.com.tw/articles/view/58468?utm_source=dable'
html = requests.get(url)
#html.encoding='utf-8'
#print(html.text)
soup = BeautifulSoup(html.text, 'html5lib')
for e in soup.select('p'):
    print(e.text)
```

#### 7.4.1 import 模組
- 語法1: import 套件(模組)名稱
    - 使用函式時需用全名： 模組名.函式名
- 語法2: from 套件(模組)名稱 import 函式名稱

    - import 套件(模組)名稱 as 別名
