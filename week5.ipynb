### 串列長度: len(串列)

```
len(scores)
```

```
# 順著印 0 ~ n-1
for i in range(len(scores)):
  print(f'scores[{i}]={scores[i]}')
```

```
# 倒著印 n-1 ~ 0, 每次減1 (但最後一個不算, 故要寫成(n-1, -1, -1))
for i in range(len(scores)-1, -1, -1):
  print(scores[i], end=' ')
```

```
# 倒著印，使用負索引
for i in range(-1, -len(scores)-1, -1):
  print(scores[i], end=' ')
```

### insert() 方法: 插入元素至串列特定位置
* 串列.insert(位置, 元素值)
  - 如果位置值大於或等於串列元素個數，將如同 append() 方法一樣將串列元素加
在最後面。
  - 位置值也可以是負值，表示由串列的最後向前推算，「-1」表示最後一個元素，
「-2」表示倒數第二個元素，依此類推。

```
print(scores)
scores.insert(len(scores), 'here')
print(scores)
scores.insert(0, 'there')
print(scores)
scores.insert(-1, '不是最後')
print(scores)
```

```
scores.insert(600, 'oh!')
scores
```

### practice 3. 設計一個輸入成績的程式
* 將學生成績存入串列做為串列元素，如果輸入「-1」表示成績輸入結束，最後顯示班上總成績及平均成績。

```
scores = []
while True:
  s = int(input('學生成績: '))
  if s==-1: break
  scores.append(s)
print(f'共有{len(scores)}位學生.')
print(f'總成績:{sum(scores)}分, 平均成績:{sum(scores)/len(scores):.2f}分.')
```

### practice 4. 請設計一個輸入不喜歡的水果刪除程式，並顯示剩下的水果，如果輸入 Enter 就結束輸入.
- 一開始的水果fruits=['banana', 'apple', 'orange', 'pineapple', 'watermelon']

```
fruits=['banana', 'apple', 'orange', 'pineapple', 'watermelon']
while True:
    print(f'目前的水果:{fruits}')
    f = input('輸入不喜歡的水果: ')
    if f=='': break
    if f in fruits:
       fruits.remove(f)
    else:
       print(f'{f}不在串列中.')
```

###  由大到小排序
- 首先: 串列.sort(), 接著: 串列.reverse()
- 或是使用

```
lst1=lst1[len(lst1)//2:]+lst1[:len(lst1)//2]  # 將後半移至前半
lst1
```

## 元組 (Tuple)
- 元組=(元素1, 元素2, ..., 元素n)

```
t1 = (3)  # 用括號不保證是tuple資料型態
print(t1)
```

```
t2 = (3,)  # 只有一個元素時要加逗號 
print(t2)
```

```
t3 = 3,  # 其實不加括號也可以，重點在逗號
print(t3)
t4 = 3, 4
print(t4)
```

```
print(type(t3))
type(t4)
```

```
t5 = ()  # 但是空元素就一定要加括號
print(t5)
print(type(t5))
```

```
t6 = tuple()   # 但是空元素的tuple也可以宣告成tuple()
print(t6)
print(type(t6))
```

```
a, b= 7, 8
print((a, b))
```

```
# 兩數交換
a, b = b, a
print((a, b))
```

```
a, b = (int(x) for x in eval(input('輸入兩個整數, 以逗號隔開:')))
print((a, b))
```

```
a, b = range(7, 9)
print((a, b))
```

```
a,b,c,d,e=range(5)
print((a,b,c,d,e))
```

```
print(range(5))
print(*range(5))
```
